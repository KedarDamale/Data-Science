# **Data Cleaning**

Data Cleaning (also known as data cleansing or data scrubbing) is the process of detecting, 
correcting, or removing errors and inconsistencies in data to improve its quality and reliability. 
This is a critical step in data preprocessing because raw data is almost always incomplete, 
inconsistent, inaccurate, or duplicated, which can mislead the machine learning model or any analysis.


🔍 Detailed Steps in Data Cleaning
| Step                            | Description                                       | Methods                           |
| ------------------------------- | ------------------------------------------------- | --------------------------------- |
| 1. **Remove duplicates**        | Eliminate repeated rows                           | `df.drop_duplicates()`            |
| 2. **Handle missing values**    | Fill or drop nulls                                | `df.fillna()`, `df.dropna()`      |
| 3. **Correct data types**       | Convert strings to int/float/date etc.            | `df.astype()`                     |
| 4. **Fix inconsistencies**      | Normalize text values                             | Convert "USA", "usa" to uppercase |
| 5. **Remove outliers**          | Drop or replace extreme values                    | IQR method, Z-score method        |
| 6. **Standardize formats**      | Consistent date/currency formats                  | Use `datetime`, regex             |
| 7. **Spell check / clean text** | Remove typos, unwanted characters                 | NLP techniques                    |
| 8. **Validate ranges**          | Ensure numeric values fall within expected ranges | Logical filters                   |

```python
# Improved One-Hot Encoding Implementation
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# Create a copy of the dataframe for encoding
df_encoded = df_encode.copy()

# Initialize the encoder
encoder = OneHotEncoder()  # Using dense matrix for easier handling

# Select columns to encode - in this case just 'os'
# You can add more columns here if needed
columns_to_encode = ['os'] #always select columns to encode like this otherwise index column that is default will get encoded too

# Fit and transform the selected columns
encoded_array = encoder.fit_transform(df_encoded[columns_to_encode]).toarray()#this return array from sparse matrix , 
                                                                              #sparse matrix is a matrix that encoder 
                                                                              # #generates basically a matrix of only 0's and 1s


"""
1.fit():Learn or calculate internal statistics/parameters from the data - does not modify the data

2.transform():Use the statistics/parameters learned in fit() to transform or modify the data.

3.fit_transform():Does above two in one go


`fit_transform()` is only useful when you're working with **one dataset** (like training data).
But in real ML workflows, you must **fit on training data** and **transform test data** separately to avoid **data leakage**.
That's why `fit()` and `transform()` are separate — to allow reusing learned parameters safely across datasets.

"""


# Get feature names for the encoded columns
feature_names = encoder.get_feature_names_out(columns_to_encode)

# Create a DataFrame with the encoded values
encoded_df = pd.DataFrame(encoded_array, columns=feature_names, index=df_encoded.index)

# Remove original columns and add encoded columns
final_df = pd.concat([df_encoded.drop(columns=columns_to_encode), encoded_df], axis=1)

# Display the shape of the final dataframe
print("Original shape:", df_encoded.shape)
print("Final shape:", final_df.shape)
print("\nNew encoded columns:", list(feature_names))
```

```python
"""
Outlier removal using 

IQR=Q3-Q1

Any data point more than or less than 1.5 * IQR than Q0 or Q4 is an outlier
"""
```

```python
df.sample(5)
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Filter out rows with outliers
df_no_outliers_iqr = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

"""
.any(axis=1) → tells Pandas to: Go row by row (axis=1),
                                If any of the columns in that row has a True value (i.e. is an outlier),
                                Then mark that entire row as having an outlier.
any is used as we are going through every column if not then we would have used df['columnname']
"""

plt.boxplot(df_no_outliers_iqr)
```

```python
"""
outlier removal using zscore

Z-score tells how many standard deviations a data point is away from the mean.

If the Z-score is:

Close to 0: the data point is near the mean.
Far from 0 (say > 3 or < -3): it's considered an outlier.

"""
```

```python
from scipy.stats import zscore

# Step 3: Compute Z-scores for all columns
z_scores = np.abs(zscore(df))

# Step 4: Filter rows where ALL column z-scores < threshold (3)
threshold = 3
df_clean = df[(z_scores < threshold).all(axis=1)] #as the condition means datapoint we used all means all datapoint are datapoint if they follw this condition
                                                  #in IQR any case condition satisfaction means an outlier and there we used any condition to trace any outlier


plt.boxplot(df_clean) #there will be some outliers left but they are okk
```

```python
"""
🔹What is Feature Scaling?

In machine learning and data preprocessing, feature scaling is the process
of normalizing or transforming features (variables) to a common scale without
distorting differences in the ranges of values.

🔹Why is it important?

Many ML algorithms (e.g., gradient descent-based, KNN, SVM, PCA) are sensitive to the scale of input features.
Features with larger numeric ranges can dominate those with smaller ranges.
It improves:
  Convergence speed of optimization algorithms.
  Model accuracy and stability.
  Distance-based computations (e.g., Euclidean distance in KNN).

🔹Types of Feature Scaling

There are two types of feature scaling: 1.Normalization
                                        2. Standardization
                                        
🔹Normalization: Normalization is a feature scaling technique that rescales 
                 feature values to a fixed range, typically 0,1 without 
                distorting the relative relationships between the data values.
                
                Techniques: Min-Max normalization (Most used)
                            Deviation normalization
                          
🔹Standardization: Standardization is a feature scaling technique that transforms
                   features to have a mean of 0 and a standard deviation of 1.
                   
                Techniques: Z score normalization (Most used)
                            IQR standardization

"""
```

```python
"""
🔹Duplicate data: Duplicate data refers to rows or 
                  records in a dataset that appear more
                  than once — they have the same values 
                  across all or specific columns.

🔹Why handling it is necessary
        -Duplicate entries inflate counts, averages, and skew distributions.
        -Machine learning models overfit or learn incorrect patterns.
        -More memory/storage, longer computation times.
        -In apps, duplicate rows cause confusion or errors in display/reports.

"""

import pandas as pd

# Sample dataset
data = {
    'ID': [101, 102, 103, 101, 104, 105, 102],
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'David', 'Eve', 'Bob'],
    'Department': ['HR', 'IT', 'Finance', 'HR', 'IT', 'HR', 'IT'],
    'Salary': [50000, 60000, 55000, 50000, 65000, 52000, 60000]
}

df = pd.DataFrame(data)
print("🔹 Original Dataset:\n", df,end='\n\n')
print(f"Are there any duplicated values and if then how much: {df.duplicated().sum()}",end='\n\n')

print('Dataframe after duplicate removal:',end='\n\n')

df.drop_duplicates(inplace=True) #Note, inplace is must here or initialize as it will return new df it wont modify in place

print(df,end='\n\n')
print(f"Are there any duplicated values now: {df.duplicated().sum()}",end='\n\n')



"""
Duplicates are counted when all the values of all columns match with something else but
we can manipulate this behaviour
df.duplicated() will return true false matrix like isnull .sum() will give the count 

df.duplicated(subset=['ID']) #this will only check duplicated data in ID column if ID has duplicates and next colukns are disticnt still it will mark it as duplicates
df.duplicated(subset=['ID','Name']) #same like above but for two columns
"""

```

```python
#to_numeric method in pandas

df = pd.DataFrame({
    'Price': ['1,000', '2000', 'N/A', '3,500.50', 'abc', '', '4200', '7,890.99', 'NaN', None, '-500']
})

# Step 1: Remove commas
df['Price_cleaned'] = df['Price'].replace(',', '', regex=True)

# Step 2: Convert to numeric, forcing errors to NaN
df['Price_numeric'] = pd.to_numeric(df['Price_cleaned'], errors='coerce')

print(df)
```

```python
#Manual function transformer

#Way 1: slower
df=pd.DataFrame({"temp":[30,32,31,30,29]})
l=[]
for i in df['temp']:
    l.append((i*9/5)+32)
    
df['temp_faran']=l
print(df.to_string(index=False))

#Way 2: faster

df = pd.DataFrame({"temp": [30, 32, 31, 30, 29]})
df['temp_faran'] = (df['temp'] * 9/5) + 32
print(df.to_string(index=False))

```

```python
#with function transformer

from sklearn.preprocessing import FunctionTransformer

def c_to_f(x): #function defined here which FunctionTransformer will transform our data according to
    return (x * 9/5) + 32

transformer = FunctionTransformer(c_to_f)
df['temp_faran_using_sklearn'] = transformer.fit_transform(df[['temp']]) #just transform instead of fit_transform will do, 2d array is required here 

df
```

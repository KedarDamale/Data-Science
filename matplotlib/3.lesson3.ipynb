# **Bar plot with plt.bar()**

```python
import matplotlib.pyplot as plt

x={
    "Mon":30,
    "Tue":34,
    "Wed":29,
    "Thu":24,
    "Fri":40,
    "Sat":37,
    "Sun":31,
    
}

#plt.bar(x.keys(),x.values()) #in barplot x values has to be categories and 7 values will be numerical
#x and y values are mandatory below values are not mandatory

plt.bar(x.keys(),x.values(),
        width=[1,1,1,1,0.1,1,1], #default width is 0.8 we can also pass in list of width for each bar
        bottom=100,#starting value of bars default 0
        align='edge',#bar alignment will be from edge if center then it will be centered
        color='black',#sets color of bars
        edgecolor='pink',#edge color of bars
        linewidth=7,#edge width
        alpha=1, #transparenecy 0-1 range
        label='Barplot',#label for legend method
        hatch='/X/'#any pattern you would like to fill in +-0X/
        )
```

```python
#horizontal plot just like bar plot but horizontal

import matplotlib.pyplot as plt

x = {
    "Mon": 30,
    "Tue": 34,
    "Wed": 29,
    "Thu": 24,
    "Fri": 40,
    "Sat": 37,
    "Sun": 31,
}

# FIX: wrap keys() and values() with list()
plt.barh(list(x.keys()), list(x.values()))#this is right instead of just passing we are converting to list which is more recommended as normal may cause issues at runtime
plt.xlabel("Value")
plt.ylabel("Day of Week")
plt.title("Horizontal Bar Chart: Value per Day")
plt.show()
```

# **Histograms with plt.hist()**

```python
"""
plt.hist() creates a histogram — a bar plot that shows how data is distributed across value ranges ("bins").

Instead of plotting individual values, it groups values into intervals and counts how many values fall into each bin.

"""
import matplotlib.pyplot as plt

data=[10,12,13,34,56,34,23,67,89,23,45,67,12,90,18,29,40,18,24,59,67,19,20,23,21,43,78]

plt.hist(data,#only data is mandotory
         bins=7, #bins means number of bins you want to create in the range default is 10 range/bins number will be the range of single bin
         range=(0,50),#this will decide the range tuple(min,max)
         density=False,#If you want the y-axis to represent probability density (so area under curve = 1):
         cumulative=False,#computes cumulative histogram and y axis will represetn that
         histtype='barstacked',#means multiple baars will be stacked #bar barstacked step and stepfilled 
         align='left', #aligns hists to left mid right
         orientation='vertical', #like bar and barh hstogram doesnt have diffent metho it has orientation attribute
         rwidth=0.6,#width of hists
         log=False,#if true then uses log scale for y axis instead of normal number series
         color='yellow',
         label="histogram",
         edgecolor='black',
         alpha=0.5,
         stacked=True #if multiple hists are plotted the nif true then stacks them on each other
         ) 
plt.show()
```

# **Pie chat with plt.pie()**

```python
# plt.pie() creates a circular pie chart, dividing a circle into wedges (slices) proportional to the input values.

# Each value in the data array represents the size of a slice.

# It is ideal for showing percentage breakdowns of a whole.
```

```python
x = [21, 18, 46, 50]

#for pie chart Values for slices (must be non-negative)

labels=["A","B","C","D"] #a label for each datapoint

plt.pie(x,#only x is mandotry 
        labels=labels,#this will give label to each datpoint
        autopct='%0.2f%%', #will disply there percentage datpoint(value)/total of values * 100
        explode=[0, 0.5, 0, 0] ,#to chip out particular piece from center to highlight basically
        colors=['red','yellow','turquoise','pink'],
        pctdistance=1, #distance of percentage label from center
        labeldistance=0.7,#label distance from center
        startangle=90, #Rotation of the chart in degrees
        shadow=True,#draws shadow so it looks more 3d
        radius=2,#radius of the pie
        frame=True#draws fremaw around pie like the axis frame
        )

plt.show()
```

# **Scatter chat with plt.scatter()**

```python
"""
plt.scatter() creates a scatter plot, which plots individual data points as dots on a 2D Cartesian plane.

Each point is positioned using its x and y values.

It is typically used to show relationships, patterns, clusters, outliers, or correlations between two variables.
"""

x=[10,1,5,6,7,8,3,2,4,5,6]
y=[2,4,5,6,7,1,2,9,7,3,10]

#x and y lists are coordinates meaning datapoint one will have (10,2)

plt.scatter(x,y, #x and y are mandatory
            s=100,#size of datapoints can provide a list for each datapoint size
            c=['red','pink','pink','red','red','pink','red','pink','pink','red','pink'],#color of dots can be array for each one can be one for all in a string
            marker='o',#marker shape c
            alpha=0.5,
            edgecolors='blue',# edge of datamarkwe only works if marker is a dot
            label='scatterplot'
            )

plt.show
```

```python
import matplotlib.pyplot as plt


classA_hours=[10,12,13,7,9,16,14,3,9,10]
classA_score=[80,90,98,75,80,76,54,67,90,76]

classB_hours=[7,15,11,9,9,6,4,3,9,16]
classB_score=[89,79,86,95,54,89,94,67,60,76]

plt.scatter(classA_hours,classA_score,label="Class A",marker='o',color ='blue')
plt.scatter(classB_hours,classB_score,label="Class B",marker='^',color ='red')

plt.grid()
plt.legend()
plt.show()

```

```python
#Syntax: plt.subplot(nrows(number of rows in canvas),ncols(number of columns in canvas),index(1 based indexing not 0))
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[10,20,15,25]

plt.subplot(1,2,1) #one row two columns in canvas and in that  1st position
plt.plot(x,y,label="line chart")
plt.title("Line chart")
plt.grid()
plt.legend()

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title("Bar plot")


plt.tight_layout()#to prevent overlap
plt.show()
```

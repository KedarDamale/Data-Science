# **plt.plot() function**

```python
#plt.plot(x,y)#mandatory parameter x being data on x axis and y beibg data on y axis

#pl.plot(x,y,color='color name or hex code',linestyle='linestyle',linewidth=int value,marker="marker around datapoint",labelname)

"""
| Parameter                 | Description                                                       |
| ------------------------- | ----------------------------------------------------------------- |
| `color` / `c`             | Color of the line (e.g., `'red'`, `'g'`, `'#ff8800'`)             |
| `linestyle` / `ls`        | Line style (`'-'`, `'--'`, `'-.'`, `':'`, `'None'`)               |
| `linewidth` / `lw`        | Thickness of the line                                             |
| `marker`                  | Marker style for points (e.g., `'o'`, `'s'`, `'^'`, `'*'`, `'x'`) |
| `label`                   | Name to use in legend                                             |
| `alpha`                   | Transparency (float from 0 to 1)                                  |
| `markersize` / `ms`       | Size of the marker                                                |
| `markeredgecolor` / `mec` | Color of marker's edge                                            |
| `markerfacecolor` / `mfc` | Fill color of marker                                              |
| `markeredgewidth` / `mew` | Width of marker‚Äôs edge                                            |
| `zorder`                  | Layering (higher values on top)                                   |

"""
```

```python
import matplotlib.pyplot as plt

x={
    "Mon":30,
    "Tue":34,
    "Wed":29,
    "Thu":24,
    "Fri":40,
    "Sat":37,
    "Sun":31,
    
}

plt.plot(x.keys(), x.values(),
         color='orange',
         linestyle=':',
         linewidth=5,
         marker='o',
         markersize=10,
         markerfacecolor='blue',
         markeredgecolor='black',
         markeredgewidth=1,
         alpha=0.9,
         label='Sample Line',
         zorder=3)

plt.show()
```

# **plt.xlabel() and plt.ylabel() function**

```python
# a string as a label is mandatory

# Optional Parameters (kwargs)

# These optional parameters are passed to control **text styling and positioning**.

# üìê Position & Alignment

# | Parameter  | Type  | Description                                                                                        |
# | ---------- | ----- | -------------------------------------------------------------------------------------------------- |
# | `loc`      | str   | `'left'`, `'center'`, `'right'` (for `xlabel`) <br> `'bottom'`, `'center'`, `'top'` (for `ylabel`) |
# | `labelpad` | float | Padding between label and axis (in points)                                                         |
# | `x` / `y`  | float | Manual positioning: fraction of axis (0.0 to 1.0)                                                  |

# üé® Text Style

# | Parameter    | Type         | Description                                            |
# | ------------ | ------------ | ------------------------------------------------------ |
# | `fontsize`   | int or str   | Size of the text (`12`, `'large'`, etc.)               |
# | `color`      | str          | Color of the label text                                |
# | `fontweight` | str          | `'normal'`, `'bold'`                                   |
# | `fontstyle`  | str          | `'normal'`, `'italic'`, `'oblique'`                    |
# | `rotation`   | float or str | Rotate label text (`90`, `'horizontal'`, `'vertical'`) |
# | `fontname`   | str          | Font family (e.g., `'Arial'`, `'Times New Roman'`)     |
# | `alpha`      | float        | Transparency (0 to 1)                                  |
```

```python
x={
    "Mon":30,
    "Tue":34,
    "Wed":29,
    "Thu":24,
    "Fri":40,
    "Sat":37,
    "Sun":31,
    
}
plt.plot(x.keys(),x.values())

plt.xlabel("Days",
           fontsize=12, 
           color='blue', 
           fontweight='bold',
           labelpad=15, 
           loc='right', 
           alpha=0.8)

plt.ylabel("Sales",
           fontsize=12, 
           color='blue', 
           fontweight='bold',
           labelpad=15, 
           loc='top', 
           alpha=0.8)
           


plt.show()
```

# **plt.title() function**

```python
# plt.title("Your Title", **kwargs) title is mandatory parameter

"""
OPtional params

| Parameter | Type    | Description                                                     |
| --------- | ------- | --------------------------------------------------------------- |
| `loc`     | `str`   | Horizontal alignment: `'left'`, `'center'` (default), `'right'` |
| `pad`     | `float` | Padding between title and top of axes (in **points**)           |
| `x`       | `float` | Manual horizontal position (0.0 = far left, 1.0 = far right)    |
| `y`       | `float` | Manual vertical position (default \~1.0)                        |

| Parameter         | Type           | Description                                          |
| ----------------- | -------------- | ---------------------------------------------------- |
| `fontsize`        | `int` or `str` | Size of the title (`14`, `'large'`, etc.)            |
| `color`           | `str`          | Color of the title text                              |
| `fontweight`      | `str`          | `'normal'`, `'bold'`, etc.                           |
| `fontstyle`       | `str`          | `'normal'`, `'italic'`, `'oblique'`                  |
| `fontname`        | `str`          | Font family (e.g., `'Arial'`, `'Times New Roman'`)   |
| `alpha`           | `float`        | Transparency of the title (0 to 1)                   |
| `backgroundcolor` | `str`          | Background color behind the title text               |
| `rotation`        | `float`        | Rotation of title text in degrees (rarely used here) |


"""
```

```python
x={
    "Mon":30,
    "Tue":34,
    "Wed":29,
    "Thu":24,
    "Fri":40,
    "Sat":37,
    "Sun":31,
    
}
plt.plot(x.keys(),x.values())

plt.title("Sales Data",
          fontsize=16,
          color='white',
          fontweight='bold',
          fontstyle='italic',
          loc='left',
          pad=20,
          x=1,
          y=1,
          backgroundcolor="#dd0000",
          alpha=0.9)

plt.show()
```

# **plt.legend() function**

```python
"""
plt.plot(x, y, label="Speed")  # attach label to the plot
plt.legend()                   # show the legend box

optional parameters
| Parameter        | Type       | Description                                                                  |
| ---------------- | ---------- | ---------------------------------------------------------------------------- |
| `loc`            | str or int | Position of legend: `'best'` (auto), `'upper left'`, `'lower right'`, etc.   |
| `bbox_to_anchor` | tuple      | Manual (x, y) positioning outside axes: e.g., `(1, 1)` places legend outside |
| `ncol`           | int        | Number of columns (e.g., 2 for side-by-side entries)                         |
| `frameon`        | bool       | Whether to draw a border box around legend                                   |
| `borderpad`      | float      | Padding inside legend box                                                    |
| `labelspacing`   | float      | Vertical space between labels                                                |
| `handlelength`   | float      | Length of line before label                                                  |
| `handletextpad`  | float      | Space between line and text                                                  |
| `columnspacing`  | float      | Horizontal space between columns                                             |
| `title`          | str        | Optional title for the legend box                                            |

| Parameter        | Type       | Description                              |
| ---------------- | ---------- | ---------------------------------------- |
| `fontsize`       | str or int | Text size (`'small'`, `12`, `'x-large'`) |
| `shadow`         | bool       | Adds shadow behind legend box            |
| `fancybox`       | bool       | Rounded corners on the legend box        |
| `edgecolor`      | str        | Border color                             |
| `facecolor`      | str        | Background color of the legend box       |
| `title_fontsize` | str/int    | Font size for the title                  |
| `framealpha`     | float      | Transparency of legend box (0.0 to 1.0)  |



"""
```

```python
import matplotlib.pyplot as plt
x={
    "Mon":30,
    "Tue":34,
    "Wed":29,
    "Thu":24,
    "Fri":40,
    "Sat":37,
    "Sun":31,
    
}
plt.plot(x.keys(),x.values())

plt.legend(loc='upper left',
           fontsize='medium',
           title='Legend Box',
           title_fontsize='large',
           frameon=True,
           shadow=True,
           fancybox=True,
           framealpha=0.8,
           facecolor="#9b1717",
           edgecolor='black',
           ncol=1,
           labelspacing=0.5,
           handlelength=2.5,
           handletextpad=0.8,
           borderpad=1.2,
           columnspacing=1.0)



plt.show()
```

```python
"""
plt.xticks(ticks=None, labels=None, **kwargs)
plt.yticks(ticks=None, labels=None, **kwargs)

| Parameter  | Type          | Description                                                   |
| ---------- | ------------- | ------------------------------------------------------------- |
| `ticks`    | list or array | The **positions** on the axis where you want ticks to appear  |
| `labels`   | list of str   | The **labels to show** at those tick positions                |
| `**kwargs` | styling       | Optional text styling (`fontsize`, `rotation`, `color`, etc.) |

| Parameter    | Type      | Description                                           |
| ------------ | --------- | ----------------------------------------------------- |
| `fontsize`   | int/str   | Tick label size                                       |
| `rotation`   | float/int | Rotate text (e.g. `90`, `'vertical'`, `'horizontal'`) |
| `color`      | str       | Color of tick text                                    |
| `fontweight` | str       | `'bold'`, `'normal'`                                  |
| `fontname`   | str       | Font family (e.g., `'Arial'`)                         |


"""
```

# **plt.xslim() and plt.yslim functions**

```python
#They set or return the limits (min and max values) for the X-axis and Y-axis, respectively.


"""
plt.xlim(left=None, right=None)
plt.ylim(bottom=None, top=None)

| Parameter (xlim) | Type         | Description          |
| ---------------- | ------------ | -------------------- |
| `left`           | float or int | Minimum X-axis value |
| `right`          | float or int | Maximum X-axis value |

"""
```

```python
# These are extremely useful when:

# You want to zoom into a specific region

# You want to standardize plot limits across multiple graphs

# You're debugging and want to exclude outliers or empty space



import matplotlib.pyplot as plt

x={
    "Mon":30,
    "Tue":34,
    "Wed":29,
    "Thu":24,
    "Fri":40,
    "Sat":37,
    "Sun":31,
    
}
plt.plot(x.keys(),x.values())

plt.xlim(0, 10)
plt.ylim(0, 60)

plt.title("Zoomed Plot")
plt.grid(True)
plt.show()
```
